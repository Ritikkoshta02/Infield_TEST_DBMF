```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# Kaggle paths
episodes_path = '/kaggle/input/ritik-mtp/episode_numbers.txt'
footprints_path = '/kaggle/input/ritik-mtp/footprints.txt'

# Load data from text files
episodes = np.loadtxt(episodes_path)
footprints = np.loadtxt(footprints_path)

# Group size
group_size = 100

# Compute grouped averages
episodes_avg = np.mean(episodes.reshape(-1, group_size), axis=1)/100.0
footprints_avg = np.mean(footprints.reshape(-1, group_size), axis=1)

# Smooth the averaged footprints using Gaussian filter
footprints_smoothed = gaussian_filter1d(footprints_avg, sigma=60)

# Set font to Times New Roman, bold and size 20
# plt.rcParams["font.family"] = ""
plt.rcParams["font.weight"] = "bold"
plt.rcParams["font.size"] = 18

# Plotting grouped averages and smoothed version
plt.figure(figsize=(12, 6))
plt.plot(episodes_avg, footprints_avg, linestyle='-', linewidth=2, color='blue', markersize=4, label='Actual Footprints')
plt.plot(episodes_avg, footprints_smoothed, linestyle='-', linewidth=4, color='orange', markersize=4, label='Moving Average')

# Labels and title
plt.xlabel('Number of Epochs', fontsize=20, fontweight = 'bold')
plt.ylabel('Footprints Covered per Epoch', fontsize=20, fontweight = 'bold')
# plt.title('Footprints vs Number of Episodes (Grouped Avg & Smoothed)', fontsize=16)
plt.legend()
# plt.grid(True)

plt.tight_layout()
plt.savefig('footprints_vs_epochs.pdf', format='pdf')
plt.show()
```
